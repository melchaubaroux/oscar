```python
import pandas as pd

male=pd.read_csv('./data/oscar_age_male_raw.csv', skipinitialspace=True, index_col='Index')
female=pd.read_csv('./data/oscar_age_female_raw.csv', skipinitialspace=True, index_col='Index')
```

```python
def format(*args):
    
    for data in args:

        # Réinitialiser l'index
        data.reset_index(drop=True, inplace=True)
        
        # Ajouter 1 à la colonne 'Year'
        if 'Year' in data.columns:
            data['Year'] -= 1
        
        # Appliquer les remplacements pour les valeurs de type chaîne
        data.replace(['  ','"',"\t"], '', regex=True, inplace=True)
      
        # Convertir les noms de colonnes en minuscules
        data.columns = data.columns.str.lower()
```

```python
latest_female_oscars = pd.DataFrame({
    'year': [2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'age': [28, 60, 44, 50, 63, 44, 60],
    'name': ['Emma Stone', 'Frances McDormand', 'Olivia Colman', 'Renée Zellweger', 'Frances McDormand', 'Jessica Chastain', 'Michelle Yeoh'],
    'movie': ['La La Land', 'Three Billboards Outside Ebbing, Missouri', 'The Favourite', 'Judy', 'Nomadland', 'The Eyes of Tammy Faye', 'Everything Everywhere All at Once']
})
female = pd.concat([female, latest_female_oscars])
female
```

```python
latest_male_oscars = pd.DataFrame({
    'year': [2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'age': [41, 59, 37, 45, 83, 53, 54],
    'name': ['Casey Affleck', 'Gary Oldman', 'Rami Malek', 'Joaquin Phoenix', 'Anthony Hopkins', 'Will Smith', 'Brendan Fraser'],
    'movie': ['Manchester by the Sea', 'Darkest Hour', 'Bohemian Rhapsody', 'Joker', 'The Father', 'King Richard', 'The Whale']
})
male = pd.concat([male, latest_male_oscars])
male
```

```python
format(male,female)

female['genre']='F'
male['genre']='M'

male.to_csv("./oscar_age_male_cleaned.csv")
female.to_csv("./oscar_age_female_cleaned.csv")

df_concat = pd.concat([male,female]).sort_values(by='year')
df_concat.to_csv('./data/oscar_age_cleaned.csv', index=False)
```
